# DBSCAN Clustering with Iris Dataset

## Import Libraries

```python
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
import seaborn as sns
import matplotlib.pyplot as plt
```

## Load Dataset

```python
iris = datasets.load_iris()
X = iris.data
```

## Use StandardScaler and Clustering

Create an instance of the StandardScaler class in scikit-learn, which can be used to preprocess the data by removing outliers and scaling it to have zero mean and unit variance. To use the scaler, create an object named std_scl or any other variable name of our choice and call the fit_transform method on the input data to obtain the standardized version of the data.

```python
std_slc = StandardScaler()
X_std = std_slc.fit_transform(X)
```

The DBSCAN model is utilized, and the data is trained by using the output of standard scaling. After training, the resulting clusters are predicted and stored in a pandas DataFrame.

```python
model = DBSCAN()

model = model.fit(X_std)

clusters = model.fit_predict(X_std)
```

## Visualising the clusters

### PCA dimensionality reduction

```python
from sklearn import decomposition

pca = decomposition.PCA(n_components=2)
pca.fit(iris.data)
Z = pca.transform(X)

fig = plt.figure(figsize=(10,10)) 
ax = fig.add_subplot(111)
scatter = ax.scatter(Z[:,0],Z[:,1], c=clusters,s=50)
ax.set_title("DBSCAN Clustering\nVisualization with 2 PCA")
ax.set_xlabel("Z0")
ax.set_ylabel("Z1")
# plt.colorbar(scatter)
plt.show()
```
